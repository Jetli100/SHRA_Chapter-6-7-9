```python
# Note: The codes were originally created by Prof. Jack Baker in the MATLAB

import matplotlib.pyplot as plt
import numpy as np
import math
import os
%matplotlib inline
```

```python
x = np.arange(0.0, 2.01, 0.01)
p = 1 - np.exp(-x)
```

```python
# Fig 6.2
plt.figure(figsize=(9, 6.5))
plt.plot(x, p, 'b-', [0, 2], [1, 1],'k--', [0, 2], [0, 2],'k-' )
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.xticks(np.arange(0, 2.5, step = 0.5))
plt.yticks(np.arange(0, 2.5, step = 0.5))
plt.xlabel('Expected number of occurences, $\lambda$t', fontsize = 12)
plt.ylabel('P(at least one occurence in time t), 1-$e^{-\lambda t}$', fontsize = 12)
```
